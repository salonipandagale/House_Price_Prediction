```python
#now lets make list of categorical features with missing values
#then we will replace all categorical missing values

feature_nan = [feature for feature in dataset.columns if (dataset[feature].isnull().sum()>1) and dataset[feature].dtype=='O' ]

for feature in feature_nan:
    print("{} - {}% missing values".format(feature,np.round(dataset[feature].isnull().mean(),4)))
```

```python
#lets replace missing value with "Missing laebl"

def replace_missing(dataset, feature_nan) :
  data = dataset.copy()
  data[feature_nan] = data[feature_nan].fillna("Missing")
  return data

dataset = replace_missing(dataset, feature_nan)
#now we will se if Nan values are replaced or not
print("{} and missing number of values {}".format(feature_nan, dataset[feature_nan].isnull().sum()))
```

Lets start process for replacing NaN or null values in numerical features

```python
numerical_with_nan = [feature for feature in dataset.columns if dataset[feature].isnull().sum()>1 and dataset[feature].dtype != "O"]
print(dataset[numerical_with_nan].isnull().sum())
```

```python

def replace_NaN_Numeric(dataset, numerical_with_nan):
    data = dataset.copy()

    for feature in numerical_with_nan:
        if data[feature].notnull().sum() > 0:   # only if some values exist
            median_value = data[feature].median()
            data[feature] = data[feature].fillna(median_value)

    return data

dataset = replace_NaN_Numeric(dataset, numerical_with_nan)
#now we will se if Nan values are replaced or not
print("{} and missing number of values {}".format(numerical_with_nan, dataset[numerical_with_nan].isnull().sum()))
```

```python
###we are creating time variables
for feature in ['Year Built',	'Year Remod/Add',	'Garage Yr Blt']:
    dataset[feature]=dataset['Yr Sold']-dataset[feature]

dataset.head()
```

```python
dataset[['Year Built',	'Year Remod/Add',	'Garage Yr Blt']].head()
```

```python
###converting skewed features in log normal distribution
import numpy as np
num_features=['Lot Frontage','Lot Area','1st Flr SF','Gr Liv Area','SalePrice']
for feature in num_features:
    dataset[feature]=np.log(dataset[feature])

dataset.head()
```

```python
###Rare Categorical Feature
##means category that is present less than 1 percent
```

```python
categorical_features=[feature for feature in dataset.columns if dataset[feature].dtype=='O']
categorical_features
```

```python
#lets remove rare catego

for feature in categorical_features :
  temp = dataset.groupby(feature)['SalePrice'].count()/len(dataset)
  temp_df = temp[temp>0.01].index
  dataset[feature] = np.where(dataset[feature].isin([temp_df]), dataset[feature], 'Rare_Var')

dataset.head()
```

```python
#groupby([feature]) - >  Groups data by category.
#['SalePrice'].mean() -> Calculates average SalePrice per category.
#Sort -> Sort from lowest mean to highest:
#.index -> just extract categories name in sorted order

for feature in categorical_features :
  label_ordered = dataset.groupby(feature)['SalePrice'].mean().sort_values().index
  label_ordered={k:i for i,k in enumerate(label_ordered,0)}
  dataset[feature]=dataset[feature].map(label_ordered)

dataset.head()
```

### **##feature scaling**

```python
feature_scale = [feature for feature in dataset.columns if feature not in ['PID', 'SalePrice']]
print(len(feature_scale))
print(feature_scale)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(dataset[feature_scale])
###transform the train, test set
sc = scaler.transform(dataset[feature_scale])
```

```python
data = pd.concat([dataset[['PID','SalePrice']].reset_index(drop=True),
                  pd.DataFrame(sc,columns=feature_scale)],axis=1)
data.head()
```
